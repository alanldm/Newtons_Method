$x^3+1=0$

$3x^2=0$

$x_{n+1}=x_n - \frac{f(x)}{f'(x)}$

```python
import numpy as np
import plotly.express as px
import cmath
```

```python
def f(x):
  return x**3 + 1
```

```python
def df(x):
  return 3*x**2
```

```python
size = 1001
loops = 50
val = int(size/2)

grade = np.zeros((size, size), dtype=complex)

for y in range(val, -(val+1), -1):
  for x in range(-val, val+1):
    x0 = complex((x), (y))
    for l in range(loops):
      if(x0 != complex(0,0)):
        x0 = x0 - f(x0)/df(x0)
    grade[val-y][x+val] = x0

grade
```

```python
import matplotlib.pyplot as plt

classes = np.zeros((size, size))

for i in range(len(classes)):
  for k in range(len(classes)):
    if grade[i][k].real == -1:
      classes[i][k] = 1
    elif grade[i][k].imag > 0:
      classes[i][k] = 2
    else:
      classes[i][k] = 3

fig, ax = plt.subplots(figsize=(12,12))

ax.imshow(classes, cmap='Paired')

plt.axvline(val, c='black', ls='-')

plt.axhline(val, c='black', ls='-')

plt.show()
```
